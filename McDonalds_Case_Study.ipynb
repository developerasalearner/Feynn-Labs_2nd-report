# Encoding Part

```python
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

object_columns = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting', 'Like',
       'VisitFrequency', 'Gender']

for column in object_columns:
    data[column] = le.fit_transform(data[column])
```

```python
plt.figure(figsize=(25,20))
cor=data.corr()
sns.heatmap(cor,annot=True)
plt.show()
```

# Application of Different Clustering Algorithms

K Means

```python
from sklearn.cluster import KMeans
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(data)
cluster_labels = kmeans.labels_
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['VisitFrequency'], c=cluster_labels, cmap='viridis', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Visit Frequency (Daily)')
plt.title('K-means Clustering of McDonald\'s Data')
plt.colorbar(label='Cluster')
plt.show()
```

Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
hierarchical_clustering = AgglomerativeClustering(n_clusters=num_clusters)
cluster_labels = hierarchical_clustering.fit_predict(data)
```

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt

linkage_matrix = hierarchy.linkage(data, method='ward')

plt.figure(figsize=(12, 6))
hierarchy.dendrogram(linkage_matrix, truncate_mode='lastp', p=20)
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
epsilon = 0.5
min_samples = 5
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
cluster_labels = dbscan.fit_predict(data)
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(data['Age'], data['VisitFrequency'], c=cluster_labels, cmap='viridis', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Visit Frequency (Weekly)')
plt.title('DBSCAN Clustering of McDonald\'s Data')

plt.colorbar(label='Cluster')

plt.show()
```
